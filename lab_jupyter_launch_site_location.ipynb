```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import radians, cos, sin, asin, sqrt

def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance in kilometers between two points
    on the Earth (specified in decimal degrees)
    """
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371  # Radius of Earth in km
    return c * r
landing_sites = pd.DataFrame({
    'Name': ['Landing Zone 1', 'Landing Zone 2'],
    'Lat': [28.5623, 34.632],
    'Long': [-80.5774, -120.6108]
})
import folium

map_proximity = folium.Map(location=[37.0, -95.0], zoom_start=4)
import folium

map_proximity = folium.Map(location=[37.0, -95.0], zoom_start=4)
# Draw lines from each launch site to each landing site
for idx_l, launch in spacex_df.iterrows():
    for idx_ls, landing in landing_sites.iterrows():
        distance = haversine(launch['Long'], launch['Lat'], landing['Long'], landing['Lat'])
        
        # Draw a line on the map
        folium.PolyLine(
            locations=[[launch['Lat'], launch['Long']], [landing['Lat'], landing['Long']]],
            color='green' if launch['class']==1 else 'red',
            weight=2,
            popup=f"{distance:.2f} km"
        ).add_to(map_proximity)
map_proximity
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians

# --- Fonction pour calculer la distance ---
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # rayon approximatif de la Terre en km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance = R * c
    return distance

# --- Coordonnes du site de lancement (Cape Canaveral) ---
launch_site_lat = 28.5623
launch_site_lon = -80.5774

# --- Coordonnes du point le plus proche de la cte ---
coastline_lat = 28.56367
coastline_lon = -80.57163

# --- Calcul de la distance ---
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon,
                                        coastline_lat, coastline_lon)

# --- Crer la carte ---
m = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=12)

# Ajouter Marker pour le launch site
folium.Marker([launch_site_lat, launch_site_lon], popup="Launch Site").add_to(m)

# Ajouter Marker pour le point sur la cte
folium.Marker([coastline_lat, coastline_lon],
              popup="Closest Coastline",
              icon=folium.Icon(color='red')).add_to(m)

# Ajouter DivIcon pour afficher la distance
distance_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html='<div style="font-size: 12pt; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance_coastline),
    )
)
distance_marker.add_to(m)

# Tracer une ligne entre launch site et coastline
folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]],
    color='blue',
    weight=2
).add_to(m)

# Afficher la carte
m
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
import folium
from folium.features import DivIcon
from math import radians, sin, cos, sqrt, atan2

# Fonction pour calculer la distance entre 2 points GPS
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # rayon de la Terre en km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# Coordonnes du site de lancement
launch_site_lat = 28.5623
launch_site_lon = -80.5774

# Coordonnes du point proche (ville, autoroute, voie ferre)
proximity_lat = 28.5675
proximity_lon = -80.5860

# Calcul de la distance
distance = calculate_distance(launch_site_lat, launch_site_lon, proximity_lat, proximity_lon)

# Crer la carte centre sur le site de lancement
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=13)

# Ajouter le marqueur du site de lancement
folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    popup="Launch Site",
    icon=folium.Icon(color='red')
).add_to(site_map)

# Ajouter le marqueur du point proche avec la distance
folium.Marker(
    location=[proximity_lat, proximity_lon],
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html=f'<div style="font-size: 12pt; color:blue;"><b>{distance:.2f} km</b></div>'
    )
).add_to(site_map)

# Tracer la ligne entre le site et le point proche
line_coords = [[launch_site_lat, launch_site_lon], [proximity_lat, proximity_lon]]
folium.PolyLine(locations=line_coords, color='green', weight=2, opacity=0.7).add_to(site_map)

# Afficher la carte
site_map
```
